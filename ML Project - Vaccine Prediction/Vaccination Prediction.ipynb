```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Vaccine_Prediction_Dataset.csv')
df.sample(10)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
md = np.mean(df['h1n1_worry'])
md
```

```python
df['h1n1_worry'].replace(np.nan,md,inplace=True)
```

```python
mn = np.mean(df['h1n1_awareness'])
mn
```

```python
df['h1n1_awareness'].replace(np.nan,mn,inplace=True)
```

```python
from scipy import stats as st
md = st.mode(df['antiviral_medication'])
md
```

```python
md = st.mode(df['is_seas_risky'])
md
```

```python
df['is_seas_risky'].replace(np.nan,2,inplace=True)
```

```python
df.isnull().sum()
```

```python
# missing data was handled using mode
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df,columns=['age_bracket','qualification','race','sex','income_level','marital_status','housing_status','employment','census_msa'])
df.head(10)
```

```python
df.dtypes
```

```python
# data is ready
```

```python
Y = df['h1n1_vaccine']
X = df.drop(['h1n1_vaccine'],axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)

LogR = linear_model.LogisticRegression()
LogR.fit(X_train,Y_train)
LogR.score(X_train,Y_train)
```

```python
LogR.score(X_test,Y_test)
```

```python
# first result
```

```python
# applying decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(X_train,Y_train)
DTC.score(X_train,Y_train)
```

```python
DTC.score(X_test,Y_test)
```

```python
# Over-fitting
```

```python
# Attempting to reduce over-fitting by limiting depth
DTC2 = DecisionTreeClassifier(max_depth=3)
DTC2.fit(X_train,Y_train)
DTC2.score(X_train,Y_train)
```

```python
DTC2.score(X_test,Y_test)
```

```python
# this is optimum
```

```python
# applying ensemble classifications
```

```python
# bagging classifier
from sklearn.ensemble import BaggingClassifier
BC = BaggingClassifier()
BC.fit(X_train,Y_train)
BC.score(X_train,Y_train)
```

```python
BC.score(X_test,Y_test)
```

```python
# this is a case of overfitting
# limiting the number of classifiers
BC2 = BaggingClassifier(n_estimators=8)
BC2.fit(X_train,Y_train)
BC2.score(X_train,Y_train)
```

```python
BC2.score(X_test,Y_test)
```

```python
# this is the maximum obtainable optimisation
```

```python
# Ada-Boost Classifier
from sklearn.ensemble import AdaBoostClassifier
ABC = AdaBoostClassifier()
ABC.fit(X_train,Y_train)
ABC.score(X_train,Y_train)
```

```python
ABC.score(X_test,Y_test)
```

```python
# Also optimum
```

```python
# Gradient-Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier()
GBC.fit(X_train,Y_train)
GBC.score(X_train,Y_train)
```

```python
GBC.score(X_test,Y_test)
```

```python
# Optimum
```

```python
# Random-Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(X_train,Y_train)
RFC.score(X_train,Y_train)
```

```python
RFC.score(X_test,Y_test)
```

```python
# Over-fit
```

```python
# Reducing the number of classifiers
RFC2 = RandomForestClassifier(n_estimators=10)
RFC2.fit(X_train,Y_train)
RFC2.score(X_train,Y_train)
```

```python
RFC2.score(X_test,Y_test)
```
